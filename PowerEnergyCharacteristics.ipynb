```python
import pandas as pd
def convert_db(fromdb, todb,xref ):
    fromxref = xref[xref['XREF'].str.contains(fromdb + ':')]
    fromcol = fromxref['XREF'].str.split(':',expand=True)[1].copy()
    fromxref.loc[:,fromdb] = fromcol
    toxref = xref[xref['XREF'].str.contains(todb + ':')]
    tocol = toxref['XREF'].str.split(':',expand=True)[1].copy()
    toxref.loc[:,todb] = tocol
    return fromxref.merge(toxref, how='inner', on='MNX_ID',suffixes=(':'+fromdb, ':'+todb))


xref = pd.read_table('http://www.metanetx.org/cgi-bin/mnxget/mnxref/reac_xref.tsv',comment='#', 
                        names='XREF	MNX_ID'.split('\t'))
bigg2meta = convert_db('bigg','metacyc', xref)
bigg2meta
```
